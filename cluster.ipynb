```python
import re
from nltk.corpus import stopwords
# stopword = stopwords.words(['zh'])
# print(stopword)
def getChinese(context):
#     context = context.decode("utf-8") # convert context from str to unicode
    filtrate = re.compile(u'[^\u4E00-\u9FA5]') # non-Chinese unicode range
    context = filtrate.sub(r'', context) # remove all non-Chinese characters
    context = re.sub('[\W_]+', '', context)
    context = context.replace(')','')
    context = context.replace('(','')



#     context = context.encode("utf-8") # convert unicode back to str
    return context
```

```python
docs = []
vocab = set()
fp = open('./data/all_text_for_word2vec.txt', 'w')
for data in all_json_data:
    tokens = data['replaced_title_sentence'].split() + data['replaced_text_sentence'].split()
#     text = ' '.join(tokens)
    filter_tokens = []
    for token in tokens:
        filter_token  = getChinese(token)
        if filter_token != '' and len(filter_token) > 1 and filter_token not in mystopwords and '鄉' not in filter_token and '區' not in filter_token \
        and '場' not in filter_token and '縣' not in filter_token:
            filter_tokens.append(filter_token)
            vocab.add(filter_token)
    docs.append(filter_tokens)
    new_text = ' '.join(filter_tokens)
    print(new_text)
    print(new_text, file=fp)

fp.close()
```

```python
from gensim import corpora, models
from gensim.corpora import Dictionary
dictionary = Dictionary(docs)
dictionary.filter_extremes(no_below=2, no_above=0.8)
corpus = [dictionary.doc2bow(doc) for doc in docs]
vocab_length = len(dictionary)
print('Number of unique tokens: %d' % len(dictionary))
print('Number of documents: %d' % len(corpus))
```

```python
from gensim.models import LdaModel

# Set training parameters.
num_topics = 20
chunksize = 2000
passes = 20
iterations = 400
eval_every = None  # Don't evaluate model perplexity, takes too much time.

# Make a index to word dictionary.
temp = dictionary[0]  # This is only to "load" the dictionary.
id2word = dictionary.id2token

model = LdaModel(
    corpus=corpus,
    id2word=id2word,
    chunksize=chunksize,
    alpha='auto',
    eta='auto',
    iterations=iterations,
    num_topics=num_topics,
    passes=passes,
    eval_every=eval_every
)
```

```python
model.get_topics().shape
docs_lda = model[corpus[0]]
for i , c in enumerate(corpus):
    doc = docs[i]
    print(' '.join(doc))
    doc_lda = model[c]
#     print(doc_lda)
    print('topic')
    for v in doc_lda:
        topic_id, prob = v
        topic_terms = model.get_topic_terms(topic_id, 50)
        topic_text = ''
        for topic_term in topic_terms:
            term_id , term_prob = topic_term
            topic_text += dictionary[term_id] + ' '
        print(prob)
        print(topic_text)
#     break
    print('-------')

```
